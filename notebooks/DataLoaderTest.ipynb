# Import libraries and create dataloader

```python
%cd /home/minhhieu/Desktop/Hieu/Semi-supervised-img-classification
import glob
import numpy as np
import matplotlib.pyplot as plt
from dataloader import DataLoader

loader = DataLoader("data/dogs-vs-cats/train", one_hot=True, 
                    augment=True, labels_as_subdir=True, batch_size=16)
dataset = loader.get_train_dataset()
X,X_, Y = next(iter(dataset))
X = ((X + 1) * 127.5).numpy().astype(np.uint8)
X_ = ((X_+ 1) * 127.5).numpy().astype(np.uint8)

Y = np.argmax(Y.numpy(), axis=1)
```

# Supervised dataset

## Strongly augmented images

```python
fig, ax = plt.subplots(4, 4, figsize=(16, 16))
for i in range(16):
    row = i // 4
    col = i %  4
    
    ax[row][col].imshow(X_[i])
    ax[row][col].set_title(Y[i])
    
plt.show()
```

## Weakly augmented images

```python
fig, ax = plt.subplots(4, 4, figsize=(16, 16))
for i in range(16):
    row = i // 4
    col = i %  4
    
    ax[row][col].imshow(X[i])
    ax[row][col].set_title(Y[i])
    
plt.show()
```

# Unsupervised dataset

```python
loader = DataLoader("data/unlabelled", one_hot=True, augment=True, 
                    labels_as_subdir=False, batch_size=16)
dataset = loader.get_train_dataset()
X,X_ = next(iter(dataset))
X = ((X + 1) * 127.5).numpy().astype(np.uint8)
X_ = ((X_+ 1) * 127.5).numpy().astype(np.uint8)
```
